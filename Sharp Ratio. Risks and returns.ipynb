## Коэффициент Шарпа. Риск и доходность.

Работа приводит сравнение разных инвестиций, которые могут в среднем привести к схожим результатам, но с разными уровнями риска.

Используем коэффициент Шарпа, который оценивает ожидаемую доходность для двух инвестиционных возможностей и рассчитывает дополнительную доходность на единицу риска, которую инвестор может получить, выбрав одну из двух. 

В частности, рассмотрим разницу в доходах для двух инвестиций и сравнивает среднюю разницу со стандартным отклонением (в качестве меры риска) этой разницыДавайте узнаем о коэффициенте Шарпа, рассчитав его для акций двух технологических гигантов Facebook и Amazon. В качестве ориентира мы будем использовать S & P 500, который измеряет производительность 500 крупнейших акций в США.

```python
# Importing required modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Settings to produce nice plots in a Jupyter notebook
plt.style.use('fivethirtyeight')
%matplotlib inline

# Reading in the data
stock_data = pd.read_csv('stock_data.csv', parse_dates=['Date'], index_col='Date').dropna()
benchmark_data = pd.read_csv('benchmark_data.csv', parse_dates=['Date'], index_col='Date').dropna()
```

### Посмотрим на данные 
Посмотрим на данные, чтобы узнать, сколько наблюдений и переменных у нас в распоряжении.

```python
# Display summary for stock_data
print('Stocks\n')
stock_data.info()
print(stock_data.head())
# Display summary for benchmark_data
print('\nBenchmarks\n')
benchmark_data.info()
print(benchmark_data.head())
```

### Составим графики цен акций для Amazon и Facebook

Прежде чем сравнивать инвестиции в Facebook или Amazon с индексом 500 крупнейших компаний в США, давайте визуализируем данные, чтобы лучше понять, с чем мы имеем дело.

```python
from pylab import rcParams
rcParams['figure.figsize'] = 17, 10
rcParams.update({'font.size': 18})

# visualize the stock_data
stock_data.plot(subplots=True, title='Stock Data')

# summarize the stock_data
stock_data.describe()
```

### Визуализируtv  значения для S & P 500

Давайте также рассмотрим стоимость S & P 500, нашего эталона.


```python
# plot the benchmark_data
benchmark_data.plot(subplots=True, title='S&P 500')

# summarize the benchmark_data
benchmark_data.describe()
```

### Входные данные для коэффициента Шарпа: ежедневная доходность
Коэффициент Шарпа использует разницу в доходах между двумя рассматриваемыми инвестиционными возможностями.

Однако наши данные показывают историческую ценность каждой инвестиции, а не отдачу. Чтобы рассчитать доходность, нам нужно рассчитать процентное изменение стоимости от одного дня к следующему. Мы также рассмотрим сводную статистику, потому что она станет нашим входным параметром при расчете коэффициента Шарпа. 

```python
# calculate daily stock_data returns
stock_returns = stock_data.pct_change()

# plot the daily returns
stock_returns.plot()

# summarize the daily returns
stock_returns.describe()
```

### Ежедневная доходность S&P 500 

```python
# calculate daily benchmark_data returns
sp_returns = benchmark_data['S&P 500'].pct_change()

# plot the daily returns
sp_returns.plot()

# summarize the daily returns
sp_returns.describe()
```

### Расчет  прибыли для Amazon и Facebook относительно S & P 500
Далее нам необходимо рассчитать относительную доходность акций по сравнению с эталоном S & P 500. Это рассчитывается как разница в доходах между stock_returns и sp_returns за каждый день.

```python
# calculate the difference in daily returns
excess_returns = stock_returns.sub(sp_returns, axis=0)

# plot the excess_returns
excess_returns.plot()

# summarize the excess_returns
excess_returns.describe()
```

### Коэффициент Шарпа, Шаг 1: Средняя разница в дневных доходностях акций относительно S & P 500

Теперь мы можем наконец начать вычислять коэффициент Шарпа. Сначала нам нужно рассчитать среднее значение параметра extra_returns. Это говорит нам о том, насколько больше или меньше доходность инвестиций в день по сравнению с эталоном.

```python
# calculate the mean of excess_returns 
avg_excess_return = excess_returns.mean()

# plot avg_excess_returns
avg_excess_return.plot.bar(title='Mean of the Return Difference')
```

### Коэффициент Шарпа, Шаг 2: Стандартное отклонение разницы в доходах

Похоже, что между среднесуточной доходностью для Amazon и Facebook была большая разница.

Далее мы рассчитываем стандартное отклонение extra_returns. Это показывает нам уровень риска, который подразумевают инвестиции в акции по сравнению с инвестициями в S & P 500.

```python
# calculate the standard deviations
sd_excess_return = excess_returns.std()

# plot the standard deviations
sd_excess_return.plot.bar(title='Standard Deviation of the Return Difference')
```

### Собираем все вместе

Теперь нам просто нужно вычислить соотношение avg_excess_returns и sd_excess_returns. Результатом теперь является коэффициент Шарпа, который показывает, насколько больше (или меньше) возвращает рассматриваемая инвестиционная возможность на единицу риска.

Коэффициент Шарпа часто определяется в годовом исчислении путем умножения его на квадратный корень из числа периодов. Мы использовали ежедневные данные в качестве входных данных, поэтому мы будем использовать квадратный корень из числа торговых дней (5 дней, 52 недели, минус несколько выходных): √252

```python
# calculate the daily sharpe ratio
daily_sharpe_ratio = avg_excess_return.div(sd_excess_return)

# annualize the sharpe ratio
annual_factor = np.sqrt(252)
annual_sharpe_ratio = daily_sharpe_ratio.mul(annual_factor)

# plot the annualized sharpe ratio
annual_sharpe_ratio.plot(title='Annualized Sharpe Ratio: Stocks vs S&P 500')
```

### Заключение

С учетом двух коэффициентов Шарпа, на какие инвестиции мы должны пойти? В 2016 году доля Amazon по Шарпу была в два раза выше, чем у Facebook. Это означает, что инвестиции в Amazon вернулись вдвое больше, чем S & P 500 для каждой единицы риска, которую инвестор принял бы на себя. Другими словами, в условиях с поправкой на риск инвестиции в Amazon были бы более привлекательными.

Эта разница в основном была обусловлена разницей в доходах, а не в риске между Amazon и Facebook. Риск выбора Amazon вместо FB (измеряемый стандартным отклонением) был лишь немного выше, так что более высокий коэффициент Шарпа для Amazon в конечном итоге выше, в основном из-за более высокой среднесуточной доходности для Amazon.
